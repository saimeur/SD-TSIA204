Question 3

```python
G = df.gnp.mean()
I = df.invest.mean()

slope = ((df.gnp-G)*(df.invest-I)).sum()/((df.gnp-G)**2).sum()
intercept = I - slope*G

plt.scatter(df.gnp,df.invest)
plt.plot(df.gnp, slope*df.gnp + intercept, color = "red", label = "Estimation" )
plt.plot(df.gnp,df.invest, label = "Real data")
plt.xlabel("GNP")
plt.ylabel("Invest")
plt.legend(loc="upper left")
plt.show()
```

```python
beta0 = intercept
beta1 = slope

Yhat= df.gnp*beta1 + beta0

sigma2 = ((df.invest-Yhat)**2).sum()/(df.size-2)

sigma = np.sqrt(sigma2)

SSR = ((Yhat-I)**2).sum()
SST = ((df.invest-I)**2).sum()
R2 = SSR/SST

{"slope": beta1, "intercept" : beta0 ,"sdr_deviation" : sigma, "R^2_coefficient" : R2 }
```
